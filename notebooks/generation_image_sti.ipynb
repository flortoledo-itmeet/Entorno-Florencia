```python
from PIL import Image
import torch
from diffusers import StableDiffusionInpaintPipeline
```

```python
pipe = StableDiffusionInpaintPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-inpainting",
    torch_dtype=torch.float16,
)
```

```python
prompt = """Keep the original image exactly the same — face, lighting, expression, and background. Only change the hairstyle. 
Add a modern, textured quiff haircut with moderate volume on top and a low fade on the sides. 
The style should be natural and flattering for a square-shaped face. 
Avoid shaved sides or extreme mohawk styles. Make sure the new hair blends realistically with the head."""
```

```python
input_image = Image.open("../data/istockphoto-1200677760-612x612.jpg")
mask_image = Image.open("../data/mask.png")
```

```python
#image and mask_image should be PIL images.
#The mask structure is white for inpainting and black for keeping as is
image = pipe(prompt=prompt, 
            image=input_image, 
            mask_image=mask_image).images[0]
image.save("./final_image.png")
```
